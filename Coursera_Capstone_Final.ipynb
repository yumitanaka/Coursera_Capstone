__Function to explore the chosen sectors (Lee Green) and find all the 
nearby venues. The limit setting = 100 in all the sectors__

```python
LIMIT = 100

def getNearbyVenues(names, latitudes, longitudes, radius=600):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID,CLIENT_SECRET, VERSION, lat, lng,  radius, LIMIT)
            
        # make the GET request
        results1 = requests.get(url).json()["response"]["groups"][0]["items"]
       
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results1])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Sector', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## 2 - Exploratory Analysis

__Check how many venue categories are found__

```python
print('There are {} uniques categories.'.format(len(LeeGreen_venues['Venue Category'].unique())))
```

__This is to pivot the venues tables where all the rows are sectors and all columns are different venue categories__

```python
# One hot encoding
LeeGreen_onehot = pd.get_dummies(LeeGreen_venues[['Venue Category']],  prefix="", prefix_sep="")
```

```python
# Add sector column back to dataframe
LeeGreen_onehot['Sector'] = LeeGreen_venues['Sector'] 
```

```python
# Move Sector column to the first column
fixed_columns = [LeeGreen_onehot.columns[-1]] + list(LeeGreen_onehot.columns[:-1])
LeeGreen_onehot = LeeGreen_onehot[fixed_columns]
```

__Group rows by sector and by taking the mean of the frequency of occurrence of each category__

```python
LeeGreen_grouped = LeeGreen_onehot.groupby('Sector').mean().reset_index()
```

```python
LeeGreen_grouped.shape
```

__Return most common venue categories__

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# Create columns according to number of top venues
columns = ['Sector']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# Create a new dataframe
Sector_venues_sorted = pd.DataFrame(columns=columns)
Sector_venues_sorted['Sector'] = LeeGreen_grouped['Sector']

for ind in np.arange(LeeGreen_grouped.shape[0]):
    Sector_venues_sorted.iloc[ind, 1:] = return_most_common_venues(LeeGreen_grouped.iloc[ind, :], num_top_venues)
```

## 3 - Machine Learning: K-Means Clustering

```python
# Set number of clusters
kclusters = 5

LeeGreen_grouped_clustering = LeeGreen_grouped.drop('Sector', 1)

# Run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(LeeGreen_grouped_clustering)
```

__Check the labels__

```python
kmeans.labels_
```

__Add cluster labels to the actual data__

```python
LeeGreen_merged = LeeGreen_Data

# Add clustering labels
LeeGreen_merged['Cluster Labels'] = kmeans.labels_

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
LeeGreen_merged = LeeGreen_merged.join(Sector_venues_sorted.set_index('Sector'), on='Sector')

LeeGreen_merged.head() 
```

__Visualize the clusters__

```python
# Create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=15)

# Set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(LeeGreen_merged['Latitude'], LeeGreen_merged['Longitude'], LeeGreen_merged['Sector'], LeeGreen_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
